```python
path = ''  # DBファイル保存先パス
db_name = 'google.db'  # DBファイル

# db接続オブジェクトの作成
conn = sqlite3.connect(path + db_name)

# SQL(RDBを操作するための言語)を実行するためのカーソルオブジェクトを取得
cur = conn.cursor()

# SQL文を作成
# テーブルの作成
sql = 'CREATE TABLE google (id INT, name TEXT, language TEXT, stars INT);'

# SQL文を実行
cur.execute(sql)

# DBへの接続を閉じる
conn.close()
```

```python
path = ''
db_name = 'google.db'
try:
    # DB接続オブジェクトの作成
    conn = sqlite3.connect(path + db_name)
    # SQL（RDBを操作するための言語）を実行するためのカーソルオブジェクトを取得
    cur = conn.cursor()
    # SQL文の作成
    # データの挿入
    # INSERT INTO テーブル名 VALUES (列に対応したデータをカンマ区切りで);
    # 文字列データを挿入する場合は，シングルクォーテーション(')またはダブルクォーテーション(")で囲む必要がある
    sql = "INSERT INTO google (id, name, language, stars) VALUES (?, ?, ?, ?);"
    google = [
        (2, 'smccc', 'rust', 8),
        (3, 'tensorflow', 'python', 170000),
        (4, 'kaggle-python', 'python', 1500),
        (5, 'ml-cookbook', 'python', 19000),
        (6, 'or-tools', 'c++', 13000),
        (7, 'perfetto', 'c++', 5000),
        (8, 'error-prone', 'Java', 7100)
        
    ]
    # SQL文の実行
    cur.executemany(sql, google)
    # 変更をDBに反映させる
    conn.commit()
except sqlite3.Error as e:
    print('エラーが発生しました')
finally:
    # DSへの接続を閉じる
    conn.close()

```
